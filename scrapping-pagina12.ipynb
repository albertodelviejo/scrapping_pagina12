```python
def notices(soup):
    """Función que recive un objeto soup de una categoría determinada,
    y devuelve los links de todas las noticias pertenecientes a ella"""
    featured_article_fun = soup.find('section', attrs = {'class':'top-content'})
    featured_article_link = featured_article_fun.a.get('href')
    articles_list_fun = soup.find('section',attrs={'class':'list-content'}).find_all('article',attrs={'class':'article-item article-item--teaser'})
    articles_links = [article.a.get('href') for article in articles_list_fun]
    articles_links.insert(0, featured_article_link)
    return articles_links

el_pais_section = BeautifulSoup(requests.get('https://www.pagina12.com.ar/secciones/el-pais').text, 'lxml')
notices(el_pais_section)
```
